```python
import pandas as pd

# Use encoding='latin1' to handle special characters
try:
    df = pd.read_csv('Sample_Superstore.csv', encoding='latin1')
    print("Success! Data loaded with latin1 encoding.")
except:
    df = pd.read_csv('Sample_Superstore.csv', encoding='cp1252')
    print("Success! Data loaded with cp1252 encoding.")

# Now check the first few rows to make sure it looks right
print(df.head())
```

```python
df.info()
```

```python
# 1. Convert Date columns to Datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# 2. Create 'Profit Margin' (Critical for business analysis)
df['Profit Margin %'] = (df['Profit'] / df['Sales']) * 100

# 3. Create 'Days to Ship' (To analyze logistics performance)
df['Days to Ship'] = (df['Ship Date'] - df['Order Date']).dt.days

# 4. Extract Year and Month for easier filtering later
df['Order Month'] = df['Order Date'].dt.month_name()
df['Order Year'] = df['Order Date'].dt.year

# 5. Export to a new CSV for Tableau
df.to_csv('Superstore_Cleaned_Final.csv', index=False)

print("Data cleaning complete! File saved as 'Superstore_Cleaned_Final.csv'.")
print(df[['Order Date', 'Profit Margin %', 'Days to Ship']].head())
```

```python
# Grouping by Category to see average Profit Margin
print(df.groupby('Category')['Profit Margin %'].mean())
```

```python
# See if higher discounts are actually hurting profit
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.lineplot(x='Discount', y='Profit', data=df)
plt.title('Impact of Discounts on Profitability')
plt.show()
```
